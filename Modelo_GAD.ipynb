```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import plotly.express as px
import seaborn as sns
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, r2_score
```

```python
df = pd.read_csv('dataset.tsv',sep= '\t')
```

```python
df.head(10)
```

```python
x = df['renda']
y = df['idade']

```

```python
plt.scatter(x, y)
```

```python
df.columns
```

```python
df = df.drop(['nome','matrícula'], axis=1)
```

```python
df.to_csv('df.csv', index=False)
```

```python
df.head(10)
```

```python
df['gad'] = df['gad-1'] + df['gad-2'] + df['gad-3'] + df['gad-4'] + df['gad-5'] + df['gad-6'] + df['gad-7']
df.head()
```

```python
df1 = df.filter(['idade', 'renda', 'gad', 'motivação'])
df1.head()
```

```python
x = df1['gad'].values.reshape(-1, 1)
y = df1['motivação'].values.reshape(-1, 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.4, random_state=0)
```

```python
model = linear_model.LinearRegression().fit(x_train, y_train)
```

```python
alpha, beta = model.coef_[0], model.intercept_
print('Alpha = ', alpha)
print('Beta = ', beta)
```

```python
y_pred_train = model.predict(x_train)

plt.figure(figsize=(10, 7))
plt.scatter(x_train, y_train, color='blue', edgecolor='k', alpha=.75, label="Dados de treinamento")
plt.plot(x_train, y_pred_train, color='red', linewidth=1, label=r"$y_{i}$ = $\alpha+\beta x_{i}$")
plt.xlabel('soma_gad')
plt.ylabel('motivação')
plt.legend()
plt.show()
```

```python
y_pred_test = model.predict(x_test)

plt.figure(figsize=(10, 7))
plt.scatter(x_train, y_train, color='blue', edgecolor='k', alpha=.75, label="Dados de treinamento")
plt.scatter(x_test, y_test, color='green', edgecolor='k', alpha=.75, label="Dados de teste")
plt.plot(x_train, y_pred_train, color='red', linewidth=1, label=r"$y_{i}$ = $\alpha+\beta x_{i}$")
plt.xlabel('soma gad')
plt.ylabel('motivação')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(y_test, y_pred_test, color='k', edgecolor='k', alpha=.75)
plt.xlabel(r'$y_{i}$ real')
plt.ylabel(r'$y_{i}$ previsto')
plt.plot([min(y_test), max(y_test)], [min(y_pred_test), max(y_pred_test)], label='Diagonal secundária', linestyle='--', alpha=.75)

plt.legend()
plt.show()
```

```python
print(f'Score: {round(model.score(x_test, y_test),4) * 100} %')
```
